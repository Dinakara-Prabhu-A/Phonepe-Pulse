```python
import pandas as pd
import numpy as np
import json
import os
```

```python
agg_insurance = 'data/aggregated/insurance/country/india/state'
agg_insurance_list =[]
states = os.listdir(agg_insurance)
for state in states:
    years = os.listdir(agg_insurance+"/"+state)
    for year in years:
        files = os.listdir(agg_insurance+"/"+state+"/"+year)
        for file in files:
            json_file = open(agg_insurance + "/"+state + "/" +year+"/"+file,'r')
            json_data = json.load(json_file)
            for values in json_data['data']['transactionData']:
                data = dict(
                    
                    policies_purchased = values['paymentInstruments'][0]['count'],
                    premium_value = values['paymentInstruments'][0]['amount'],
                    year = year, state = state, quarter = file[0]
                )
                agg_insurance_list.append(data)
```

```python
agg_transaction = 'data/aggregated/transaction/country/india/state'
agg_transaction_list =[]
states = os.listdir(agg_transaction)
for state in states:
    years = os.listdir(agg_transaction+"/"+state)
    for year in years:
        files = os.listdir(agg_transaction+"/"+state+"/"+year)
        for file in files:
            json_file = open(agg_transaction + "/"+state + "/" +year+"/"+file,'r')
            json_data = json.load(json_file)
            for values in json_data['data']['transactionData']:
                data = dict(
                    
                    categories = values['name'],
                    all_transactions = values['paymentInstruments'][0]['count'],
                    payment_value = values['paymentInstruments'][0]['amount'],
                    year = year, state = state, quarter = file[0]

                )
                agg_transaction_list.append(data)

agg_transaction_df = pd.DataFrame(agg_transaction_list)
agg_insurance_df = pd.DataFrame(agg_insurance_list)
agg_transaction_df
```
